```python
#----------------------------------------------------- WEEK 1 --------------------------------------------------------

import csv
import random
from typing import Dict, List, Set

def spotify_songs(path)-> List[any]:
    
    '''
    This function places the lines of the file into a list.
    After it is put in a file, one of these: Oneplay, ZeroPlays or ManyPlays is added randomly to each song, so we can see whether the user likes the song or not.
    Then, the function cleans the csv file looks at the character and integers.
    Next, it replaces str(int) as int so those are real integers and the character are strings.
    It returns an easy to work with list.
    '''
    
    # Opening the file and putting everything in a list to adjust and work with the file.
    with open(path, encoding = 'utf8') as csvfile:
        reader = csv.reader(csvfile)
        # Skipping the headers
        headings = next(reader)     
        lines: List[any] = []
        format_lines: List[any] = []
        # Put the csv file into a list.
        for line in reader:     
            lines.append(line)      
        
        # Cleaning the entire list and asigning the amount of plays to each song.
        for elements in lines:
            choices=random.choice(["OnePlay","ZeroPlays","ManyPlays"]) 
            #creating a boolean might confuse songs that were just played and skiped by the user, they are not actual listens.
            elements.append(choices)        # assign one of the items: OnePlay, ZeroPlays or ManyPlays to each song
            y: int = 0
            for values in elements:     
                values = values.replace("-", '')        # Replaces every '-' with empty space
                format_values = ''
                for char in values:
                    if char.isalpha() or char.isnumeric():
                        format_values += char
                
                elements[y] = format_values
                y += 1

            format_lines.append(elements)

        for elements in format_lines:
            y: int = 0
            for strings in elements:
                if strings.isnumeric():
                    strings = int(strings) # Turns every str(int) into an int
                elements[y] = strings
                y += 1

    return(lines)

# Making a global variable.
spotify_song_set = spotify_songs('spotify-dataset.csv')

# This generates a random playlist of 50 songs from the data set. We used this mostly to run tests and not in the final code.
sample_playlist = random.choices(spotify_song_set, k=50)        

# This generates 100 playlists containing 50 songs
week1: Dict[str, str] = {}
for i in range(100):        
    week1['playlist'+ str(i+1)] = random.choices(spotify_song_set, k=50)

def get_playlist(week1: Dict[str, list]) -> List[str]:
    
    '''
    This function looks whether a playlist follows the criteria of week one: if a user has listened to three songs in a playlist and if it has not listened to three songs in the playlist.
    If so, it is appended to a new list. The function returns a list with the playlists.
    '''
    
    # Defining variables.
    count_zero: int = 0
    count_many_one: int = 0
    all_playlists: List[str] = []
         
    # Creating list with all the playlists.
    for playlist_name, playlist in week1.items():
        all_playlists.append(playlist_name)
    
    # Choosing a random playlist from the list that contains all playlists. 
    rand_playlist = random.choice(all_playlists)    
    
    # Checking if the random playlist contains 3 songs that the user has listened to already and 3 he has not heard
    # yet. If this is the case, we select this playlist. If not, we choose another random playlist and repeat the process.
    for playlist_name, playlist in week1.items():
        if playlist_name == rand_playlist:
            for songs in playlist:
                if songs[14] == 'ZeroPlays':
                    count_zero += 1
                if songs[14] == 'ManyPlays' or songs[14] == 'OnePlay': # If it is either ManyPlays or OnePlay, it means the user 'knows' the song and probabily likes it.
                    count_many_one += 1
            if count_zero >= 3 and count_many_one >= 3: # If statement to look whether it the playlist is according to the task of week 1
                valid_playlist = rand_playlist
            else:
                rand_playlist = random.choice(all_playlists)
                continue  
            return valid_playlist

def get_songs(get_playlist: Set[str]) -> None:
    
    '''
    This function prints 5 numbers out of a random playlist in which the user has listened to three songs and has not listened to three songs.
    Generating a set instead of a list will prevent the algorithm from recommending repeated songs. 
    '''
    
    # Defining variables.
    good_playlist: str = get_playlist(week1)
    list_with_songs: Set[str] = []
        
    # Making a Set with all the songs from the chosen playlist.
    for playlist_name, playlist in week1.items(): # if the name of the list with the correct playlists is the same as the random playlist, it puts the name of the playlist in a new Set
        if playlist_name == good_playlist:
            for items in playlist:
                list_with_songs.append(items[0])
    
    # Printing 5 random songs from the list of songs from the random playlist.
    for i in range(5): 
        choice = random.choice(list_with_songs)
        print(choice)

get_songs(get_playlist)
```

```python
#----------------------------------------------------- WEEK 2 --------------------------------------------------------

def spotify_songs(path) -> list:
    
    '''
    This function goes through the csv file again, similar to week 1 but it also replaces every sub-genre to its main genre: Rock, Techno or Pop. 
    Then it also looks at whether a character is a char or int. 
    At the end it converts evert str(int) to an int. It returns a list with the easy to work with songs.
    '''
    
    with open(path, encoding='utf8') as csvfile:
        reader = csv.reader(csvfile)
        # Skipping the headers
        headings = next(reader)     
        lines: List[any] = []
        format_lines: List[any] = []
        # Putting the csv file into a list
        for line in reader:
            lines.append(line)      
        
        # Replaces every sub-genre with it's main genre
        for elements in lines:
            choices=random.choice(["OnePlay","ZeroPlays","ManyPlays"])
            elements.append(choices) # Assign one of the items: OnePlay, ZeroPlays or ManyPlays to each song
            y: int = 0
            for values in elements:     
                values = values.replace("-", '')
                values = values.replace("neo mellow", 'Rock')
                values = values.replace("permanent wave", 'Rock')
                values = values.replace("indie pop", 'Rock')
                values = values.replace("celtic rock", 'Rock')
                values = values.replace("alaska indie", 'Rock')
                values = values.replace("boy band", 'Rock')
                values = values.replace("art pop", 'Pop')
                values = values.replace("australian pop", 'Pop')
                values = values.replace("barbadian pop", 'Pop')
                values = values.replace("acoustic pop", "Pop")
                values = values.replace("baroque pop", 'Pop')
                values = values.replace("canadian pop", 'Pop')
                values = values.replace("candy pop", 'Pop')
                values = values.replace("dance pop", 'Pop')
                values = values.replace("colombianPop", 'Pop') 
                values = values.replace("pop", 'Pop')
                values = values.replace("britishsoul", 'Pop') #
                values = values.replace("australiandance", 'Pop') #
                values = values.replace("latin", 'Pop') #
                values = values.replace("downtempo", 'Pop') #
                values = values.replace("canadiantemporaryrb", 'Pop') #
                values = values.replace("danishPop", 'Pop') #
                values = values.replace("hipPop", 'Pop') #
                values = values.replace("edm", "Techno")
                values = values.replace("electro", 'Techno')
                values = values.replace("electro pop", 'Techno')
                values = values.replace("tropical house", 'Techno')
                values = values.replace("big room", 'Techno') #
                values = values.replace("belgianTechno", 'Techno') #
                values = values.replace("TechnoPop", 'Techno') #
                values = values.replace("complextro", 'Techno') #
                values = values.replace("belgianTechno", 'Techno') #
                values = values.replace("Technonictrap", 'Techno') #
                values = values.replace("escaperoom", 'Techno') #

                format_values: str = ''
                for char in values:
                    if char.isalpha() or char.isnumeric():      # looks at whether a character is a char or int
                        format_values += char
               
                elements[y] = format_values
                y += 1

            format_lines.append(elements)
       
        for elements in format_lines:
            y: int = 0
            for strings in elements:
                if strings.isnumeric():
                    strings = int(strings)      # turns every str(int) into an int
                elements[y] = strings
                y += 1

    return(format_lines)

spotify_song_set = spotify_songs('spotify-dataset.csv')

#Defining different genres
genero1: str = "Rock"
genero2: str = "Pop"
genero3: str = "Techno"


def spotify_rock(genero1: str, genero2: str, genero3: str) -> List[str]:
    
    '''
    This function looks whether every song in the original playlist is form the genre 'Rock' or not.
    If that is not the case, it moves that particular song from the list. By doing this, it returns a list with only songs with the genre 'Rock'.
    '''
    
    # Copies the spotify_song_set to work with it without changing the original list
    spotify_song_set_rock: List[any] = spotify_song_set.copy()  
    
    for song in spotify_song_set_rock.copy():
        if genero1 not in song:     # if the genre 'Rock' is not an item in the song, it gets removed
            spotify_song_set_rock.remove(song)
    return spotify_song_set_rock

rock_songs = spotify_rock(genero1, genero2, genero3)


def spotify_pop(genero1: str, genero2: str, genero3: str) -> List[str]:
    
    '''
    This function looks whether every song in the original playlist is form the genre 'Pop' or not.
    If that is not the case, it moves that particular song from the list. By doing this, it returns a list with only songs with the genre 'Pop'.
    '''
    
    # copies the spotify_song_set to work with it without changing the original list
    spotify_song_set_pop: List[any] = spotify_song_set.copy()      
    
    for song in spotify_song_set_pop.copy():
        if genero2 not in song:     # if the genre 'Rock' is not an item in the song, it gets removed
            spotify_song_set_pop.remove(song)
    return spotify_song_set_pop

pop_songs = spotify_pop(genero1, genero2, genero3)


def spotify_techno(genero1: str, genero2: str, genero3: str) -> List[str]:
    
    '''
    This function looks whether every song in the original playlist is form the genre 'Techno' or not.
    If that is not the case, it moves that particular song from the list. By doing this, it returns a list with only songs with the genre 'Techno'.
    '''
    
    # copies the spotify_song_set to work with it without changing the original list
    spotify_song_set_techno = spotify_song_set.copy()     
    
    for playlist in spotify_song_set_techno.copy():
        if genero3 not in playlist:     # if the genre 'Rock' is not an item in the song, it gets removed
            spotify_song_set_techno.remove(playlist)
    return spotify_song_set_techno

techno_songs = spotify_techno(genero1, genero2, genero3)

# Get a single random playlist from spotify_songs_set
single_playlist = random.choices(spotify_song_set, k=50)        


def genres(single_playlist: List[str]) -> None:
    
    '''
    This function will take a single playlist and count the amount of songs each genre has. Then it will print if the
    user is a pop, techno or rock fan by looking at the amount of songs in the playlist compared to eachother. 
    It will lastly recommend 5 other songs to the user that belong to this genre.
    '''
    
    # Defining variables for counting
    pop_count: int = 0
    rock_count: int = 0
    techno_count: int = 0
    pop_fan = False
    rock_fan = False
    techno_fan = False

    # For loop to count how many times the genre appears in the playlist
    for song in single_playlist:        
        if song[2] == 'Pop':
            pop_count += 1
        elif song[2] == 'Rock':
            rock_count += 1
        elif song[2] == 'Techno':
            techno_count += 1

    # If one of the genres is not present in a list, its counter gets the value 1, because later on you can not divide by zero
    if pop_count == 0:      
        pop_count: int = 1
    if rock_count == 0:
        rock_count: int = 1
    if techno_count == 0:
        techno_count: int = 1

    if pop_count > rock_count and pop_count > techno_count:     # if statement to look whether pop is more present then rock and techno
        if pop_count / rock_count >= 0.1 and pop_count / techno_count >= 0.1:       # if so, it will calculate the difference and if that is more than 10% for each, its bool gets the value True
            pop_fan = True
        else:
            pop_fan = False     # If the difference is not 10% the bool pop_fan gets the value False assigned
    elif rock_count > pop_count and rock_count > techno_count:
        if rock_count / pop_count >= 0.1 and rock_count / techno_count >= 0.1:
            rock_fan = True
        else:
            rock_fan = False
    elif techno_count > pop_count and techno_count > rock_count:
        if techno_count / pop_count >= 0.1 and techno_count / rock_count >= 0.1:
            techno_fan = True
        else:
            techno_fan = False

    if pop_fan == True:     # to check if a user is a fan of: pop, rock or techno, we print a message
        print("User is a pop fan!")
    elif rock_fan == True:
        print('User is a rock fan!')
    elif techno_fan == True:
        print('User is a techno fan!')

    if pop_fan == True:

        good_playlist: str = pop_songs
        list_with_songs: List[str] = []

        # Making a list with all the songs from the chosen playlist.
        for song in good_playlist:
            list_with_songs.append(song[0])

        # Printing 5 random songs from the list of songs from the pop playlist.
        for i in range(5):
            choice = random.choice(list_with_songs)
            print(choice)

    elif rock_fan == True:      # if the user is not a pop fan, maybe it is a rock fan

        good_playlist: str = rock_songs
        list_with_songs: List[str] = []

        # Making a list with all the songs from the chosen playlist.
        for song in good_playlist:
            list_with_songs.append(song[0])

        # Printing 5 random songs from the list of songs from the random playlist.
        for i in range(5):
            choice = random.choice(list_with_songs)
            print(choice)

    elif techno_fan == True:        # if the user is not a pop or rock fan, it might be a techno fan

        good_playlist: str = techno_songs
        list_with_songs: List[str] = []

        # Making a list with all the songs from the chosen playlist.
        for song in good_playlist:
            list_with_songs.append(song[0])

        # Printing 5 random songs from the list of songs from the random playlist.
        for i in range(5):
            choice = random.choice(list_with_songs)
            print(choice)
            
genres(single_playlist)
```

```python
#----------------------------------------------------- WEEK 3 --------------------------------------------------------

def media_of_songs(path):
    
    '''
    This function calculates the average for every category of songs.
    '''
    
   # Opening the file and creating a sum of every category, initialy having the value 0
    with open(path, encoding='utf8') as file:
        dataset = csv.reader(file)
        headings = next(dataset)
        lines: List[any] = []
        format_lines: List[any] = []
        sum_happy: int = 0
        sum_party: int = 0
        sum_calming: int = 0
        sum_lounge: int = 0

        # Creating a list with all the songs
        for line in dataset:
            lines.append(line)

       
        # Taking into consideration the energy(dataset[4]), the danceability(dataset[5]), the loudness(dataset[6]) and the liveness(dataset[7]) we are creating the sum of all of them
        for i in lines:
            sum_happy += int(i[4])
            sum_party += int(i[5])
            sum_calming += int(i[6])
            sum_lounge += int(i[7])

        # we are taking the sum done previously and spliting it to the number of lines toi create the average
        media_happy = sum_happy /len(lines)
        media_party  = sum_party /len(lines)
        media_calming = sum_calming / len(lines)
        media_lounge = sum_lounge / len(lines)

    return lines, media_happy, media_party, media_calming, media_lounge

def songs_listen(number_songs, list_songs):
    
    '''
    This function checks in what category our song should be in, taking into consideration the average of one category.
    '''
    
    list_songs_random: List[any] = []
        
    # Creating a list of random songs
    for i in range(number_songs):
        list_songs_random.append(random.choice(list_songs))

        # dataset[4] = "Energy- The energy of a song - the higher the value, the more energtic" -The tempo of the song; if this value> media of all songs, this is "happy"
        # dataset[5] = "Danceability - The higher the value, the easier it is to dance to this song"; if this value > madia of all songs, this songs is "party"
        # dataset[6] = "Loudness/dB - The higher the value, the louder the song"; if this value < media of all songs, this is "calming"
        # dataset[7] = "Liveness - The higher the value, the more likely the song is a live recording"; if this value > media off all, this song is lounge
        # if our song does not go in any of this categories, we assign to it a random category
        
    for i in list_songs_random:
        copy_i = i
        if float(i[4]) > media_happy:
            i.append("happy")
        if float(i[5]) > media_party:
            i.append("party")
        if float(i[6]) <media_calming:
            i.append("calming")
        if float(i[7]) <media_lounge:
            i.append("lounge")
        if i==copy_i:
            i.append(random.choice(["happy","party","calming","lounge" ]))
    

    # Returning the list of random songs
    return list_songs_random


def number_type_suggested_songs(list_song, number_of_songs_suggested):
    '''
    This function checks the percentage that every song category has, taking into consideration the number of songs our user listens to in total and the number of songs our 
    user listens to from every category.
    '''
    number_happy: int = 0
    number_party: int = 0
    number_calming: int = 0
    number_lounge: int = 0

    
    # Counting how many songs of each type our list has
    for i in list_song:
        if "happy" in i:
            number_happy +=1
        if "party" in i:
            number_party +=1
        if "calming" in i:
            number_calming +=1
        if "lounge" in i:
            number_lounge +=1
    number_of_songs = number_lounge + number_happy + number_party + number_calming
    
    # Making a percentage of evert category taking into consideration the number of songs in each category and spliting it to the total number of songs
    percent_happy_suggested = number_happy / number_of_songs 
    percent_party_suggested = number_party / number_of_songs 
    percent_calming_suggested = number_calming / number_of_songs 
    percent_lounge_suggested = number_lounge / number_of_songs 
    
    # Decinding the number of songs from each category we should recommend to our user
    number_happy_suggested = int(percent_happy_suggested *number_of_songs_suggested +0.5)
    number_party_suggested = int(percent_party_suggested *number_of_songs_suggested + 0.5)
    number_calming_suggested = int(percent_calming_suggested *number_of_songs_suggested +0.5)
    number_lounge_suggested = int(percent_lounge_suggested *number_of_songs_suggested +0.5)

    return number_happy_suggested, number_party_suggested, number_calming_suggested, number_lounge_suggested

def list_sugested_songs(number_happy_suggested, number_party_suggested, number_calming_suggested, number_lounge_suggested, media_happy, media_party, media_calming, media_lounge, dataset):
    
    '''
    This function creates a recommendation list that has a number of songs in accordance with the percentage that each category has.
    '''
    
    happy_list: List[any] = []
    party_list: List[any] = []
    calming_list: List[any] = []
    lounge_list: List[any] = []
    
    i: int = 0
    # creating a list containing exactly the number of happy songs as we decided in the step before
    while i < int(number_happy_suggested) :
        song = random.choice(dataset)
        if float(song[4]) > media_happy:
            song.append("happy")
            happy_list.append(song)
            i +=1
            
    j: int = 0
    # creating a list containing exactly the number of party songs as we decided in the step before
    while j < int(number_party_suggested) :
        song = random.choice(dataset)
        if float(song[5]) > media_party:
            song.append("party")
            party_list.append(song)
            j +=1
    
    k: int = 0
    # creating a list containing exactly the number of calming songs as we decided in the step before
    while k < int(number_calming_suggested) :
        song = random.choice(dataset)
        if float(song[6]) > media_calming:
            song.append("calming")
            calming_list.append(song)
            k +=1
            
    l: int = 0
    # creating a list containing exactly the number of lounge songs as we decided in the step before
    while l < int(number_lounge_suggested) :
        song = random.choice(dataset)
        if float(song[7]) > media_lounge:
            song.append("lounge")
            lounge_list.append(song)
            l +=1
            
    return happy_list, party_list, calming_list, lounge_list

lines, media_happy, media_party, media_calming, media_lounge = media_of_songs("spotify-dataset.csv")
number_of_songs_listen: int = 10
list_song = songs_listen(number_of_songs_listen, lines)

number_happy_suggested, number_party_suggested, number_calming_suggested, number_lounge_suggested = number_type_suggested_songs(list_song,5)

happy_list, party_list, calming_list, lounge_list = list_sugested_songs(number_happy_suggested, number_party_suggested, number_calming_suggested, number_lounge_suggested,media_happy, media_party, media_calming, media_lounge,lines)

# creating and printing the list of suggestions
suggested_list = happy_list+ party_list+ calming_list+ lounge_list

# print(suggested_list)
for i in suggested_list:
    print(i[0])

```
